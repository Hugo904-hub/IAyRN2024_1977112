```python
# Librerías a utilizar
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
import sklearn.linear_model # contiene diferentes modelos de ML
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(horas_estudio.reshape(-1, 1), calificaciones, test_size=0.2, random_state=0);


# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo con los datos de entrenamiento
modelo.fit(X_train, y_train)

# Hacer predicciones
predicciones = modelo.predict(X_test)

# Calcular el error cuadrático medio
mse = mean_squared_error(y_test, predicciones)
print("Error Cuadrático Medio:", mse)

# Graficar el modelo de regresión lineal con las predicciones
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, predicciones, color='blue', linewidth=2)
plt.title('Regresión Lineal: Horas de Estudio vs. Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.show()
```
